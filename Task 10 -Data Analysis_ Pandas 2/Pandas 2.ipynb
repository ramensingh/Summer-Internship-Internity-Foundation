Accessing Data from Pandas Dataframe/ Reading and Writing data with pandas/ Indexing and slicing dataframe

```python
# 1.
# importing pandas library
import pandas as pd

# Initializing the nested list with Data set
player_list = [['M.S.Dhoni', 36, 75, 5428000],
            ['A.B.D Villers', 38, 74, 3428000],
            ['V.Kholi', 31, 70, 8428000],
            ['S.Smith', 34, 80, 4428000],
            ['C.Gayle', 40, 100, 4528000],
            ['J.Root', 33, 72, 7028000],
            ['K.Peterson', 42, 85, 2528000]]

# creating a pandas dataframe
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])

# data frame before slicing
df
```

```python
# 2. Slicing rows in data frame
df1 = df.iloc[0:4]

# data frame after slicing
df1
```

```python
# 3. Slicing columnss in data frame
df1 = df.iloc[:,0:2]

# data frame after slicing
df1
```

```python
# 4. Indexing pandas dataframe

# creating a pandas dataframe and indexing it using Aplhabets
df2 = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'],index=['A', 'B', 'C', 'D', 'E', 'F', 'G'])


# Displaying data frame
df2

```

```python
# 5. Applying lambda function to a column using Dataframe.assign()
# importing pandas library
import pandas as pd

# creating and initializing a list
values = [['Rohan', 455], ['Elvish', 250], ['Deepak', 495],
        ['Sai', 400], ['Radha', 350], ['Vansh', 450]]

# creating a pandas dataframe
df = pd.DataFrame(values, columns=['Name', 'Univ_Marks'])

# Applying lambda function to find percentage of
# 'Univ_Marks' column using df.assign()
df = df.assign(Percentage=lambda x: (x['Univ_Marks'] / 500 * 100))

# displaying the data frame
df
```

```python
# 6. Sorting the Data frame in Ascending order


# creating a pandas dataframe
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])

# Sorting by column 'Weight'
df.sort_values(by=['Weight'])
```

Data Aggregation

```python
# importing pandas package
import pandas as pd

# making data frame from csv file
df = pd.read_csv("nba.csv")

# printing the first 10 rows of the dataframe
df[:10]
```

```python
# Applying aggregation across all the columns
# sum and min will be found for each
# numeric type column in df dataframe

df.aggregate(['sum', 'min'])
```

```python
# importing pandas package
import pandas as pd

# making data frame from csv file
df = pd.read_csv("nba.csv")

# We are going to find aggregation for these columns
df.aggregate({"Number":['sum', 'min'],
"Age":['max', 'min'],
"Weight":['min', 'sum'],
"Salary":['sum']})
```

```python
# importing pandas module
import pandas as pd 
 
# Define a dictionary containing employee data 
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])
 
print(df, "\n\n", df1)
```

```python
# using a .concat() method
frames = [df, df1]
 
res1 = pd.concat(frames)
res1
```
